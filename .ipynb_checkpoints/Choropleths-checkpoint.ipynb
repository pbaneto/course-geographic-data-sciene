### Choropleths
Remember these are maps that display the spatial distribution of a variable encoded in a color scheme, also called palette. 

```python
import geopandas
import seaborn as sns
import pandas as pd
from pysal.lib import examples
from pysal.viz import mapclassify
import numpy as np
import matplotlib.pyplot as plt
```

```python
examples.explain('mexico')
```

```python
# To download it
mx = examples.load_example("mexico")
```

```python
mx.get_file_list()
```

```python
db = geopandas.read_file(examples.get_path("mexicojoin.shp"))
```

```python
db.info()
```

```python
# The data doesnt include crs, we need one to add baselayers`
db.crs
```

```python
db.crs = "EPSG:4326"
db.crs
```

```python
db.columns
```

```python
# Mexico states
db.plot(
    column='INEGI', 
    categorical=True,
    legend=True
)
```

```python
# Equal interval, display a continous variable

db.plot(
    column="PCGDP2000", # GCP per capita in 2000
    scheme="equal_interval", # replacement of categorical=True
    k=7, # number of colors
    cmap="YlGn",
    legend=True
)
```

```python
# Dig into how we are encoding values as colors

classi = mapclassify.EqualInterval(db["PCGDP2000"], k=7)
classi
```

```python
# The actual break points where values stop being in one class and become part of the next one
classi.bins
```

```python
f, ax = plt.subplots(1)

# Plot the kernel density estimation (KDE)
sns.kdeplot(db["PCGDP2000"], fill=True)

# Add a blue tick for every value at the bottom of the plot (rugs)
sns.rugplot(db["PCGDP2000"], alpha=0.5)

# Loop over each break point and plot a vertical red line
for cut in classi.bins:
    plt.axvline(cut, color='red', linewidth=0.75)

plt.show()
```

```python
# lets use quantiles
db.plot(
    column="PCGDP2000", 
    scheme="quantiles", 
    k=7,
    cmap="YlGn",
    legend=True
)
```

```python
classi = mapclassify.Quantiles(db["PCGDP2000"], k=7)
classi
```

```python
f, ax = plt.subplots(1)

# Plot the kernel density estimation (KDE)
sns.kdeplot(db["PCGDP2000"], fill=True)

# Add a blue tick for every value at the bottom of the plot (rugs)
sns.rugplot(db["PCGDP2000"], alpha=0.5)

# Loop over each break point and plot a vertical red line
for cut in classi.bins:
    plt.axvline(cut, color='red', linewidth=0.75)

plt.show()
```

```python
# Fisher Jenks
db.plot(
    column="PCGDP2000", 
    scheme="fisher_jenks", 
    k=7,
    cmap="YlGn",
    legend=True
)
```

```python
classi = mapclassify.FisherJenks(db["PCGDP2000"], k=7)
classi
```

```python
f, ax = plt.subplots(1)

# Plot the kernel density estimation (KDE)
sns.kdeplot(db["PCGDP2000"], fill=True)

# Add a blue tick for every value at the bottom of the plot (rugs)
sns.rugplot(db["PCGDP2000"], alpha=0.5)

# Loop over each break point and plot a vertical red line
for cut in classi.bins:
    plt.axvline(cut, color='red', linewidth=0.75)

plt.show()
```

## Zoom into the map

```python
db.plot(
    column="PCGDP2000", 
    scheme="quantiles", 
    k=7,
    cmap="YlGn",
    legend=False
)
```

```python
f, ax = plt.subplots(1)

# Draw the choropleth
db.plot(
    column="PCGDP2000", 
    scheme="quantiles", 
    k=7,
    cmap="YlGn",
    legend=False,
    ax=ax
)
# Redimensionate X and Y axes to desired bounds
ax.set_ylim(18, 21)
ax.set_xlim(-102, -97)
```

```python
# Includes all the polygons that have at least some are within the specified range
subset = db.cx[-102:-97, 18:21]
subset.plot()
```

```python
subset.plot(
    column="PCGDP2000", 
    scheme="quantiles", 
    k=7,
    cmap="YlGn",
    legend=False, 
)

plt.show()
```

# Do-it-Yourself

```python
import pandas as pd
import geopandas
```

```python
lsoas = geopandas.read_file("data/Access_to_Healthy_Assets_and_Hazards_AHAH/Local_Authority_Districts/E08000012/shapefiles/E08000012.shp")
```

```python
ahah_data = pd.read_csv("data/Access_to_Healthy_Assets_and_Hazards_AHAH/Local_Authority_Districts/E08000012/tables/E08000012.csv")
```

```python
ahah_data.head()
```

```python
ahah = lsoas.join(ahah_data.set_index("lsoa11cd"), on="lsoa11cd")
```

```python
ahah.columns
```

## Task I: AHAH Choropleths

```python
ahah.plot(
    column="ahah", 
    scheme="equal_interval", 
    k=5,
    cmap="YlGn",
    legend=True
)
```

```python
classi = mapclassify.EqualInterval(ahah["ahah"], k=5)
classi
```

```python
f, ax = plt.subplots(1)

# Plot the kernel density estimation (KDE)
sns.kdeplot(ahah["ahah"], fill=True)

# Add a blue tick for every value at the bottom of the plot (rugs)
sns.rugplot(ahah["ahah"], alpha=0.5)

# Loop over each break point and plot a vertical red line
for cut in classi.bins:
    plt.axvline(cut, color='red', linewidth=0.75)

plt.show()
```

```python
ahah.plot(
    column="ahah", 
    scheme="quantiles", 
    k=5,
    cmap="YlGn",
    legend=True
)

classi = mapclassify.Quantiles(ahah["ahah"], k=5)
classi
```

```python
f, ax = plt.subplots(1)

# Plot the kernel density estimation (KDE)
sns.kdeplot(ahah["ahah"], fill=True)

# Add a blue tick for every value at the bottom of the plot (rugs)
sns.rugplot(ahah["ahah"], alpha=0.5)

# Loop over each break point and plot a vertical red line
for cut in classi.bins:
    plt.axvline(cut, color='red', linewidth=0.75)

plt.show()
```

```python
ahah.plot(
    column="ahah", 
    scheme="fisher_jenks", 
    k=5,
    cmap="YlGn",
    legend=True
)

classi = mapclassify.FisherJenks(ahah["ahah"], k=5)
classi
```

```python
f, ax = plt.subplots(1)

# Plot the kernel density estimation (KDE)
sns.kdeplot(ahah["ahah"], fill=True)

# Add a blue tick for every value at the bottom of the plot (rugs)
sns.rugplot(ahah["ahah"], alpha=0.5)

# Loop over each break point and plot a vertical red line
for cut in classi.bins:
    plt.axvline(cut, color='red', linewidth=0.75)

plt.show()
```

```python
avg_ahah = ahah['ahah'].mean()
avg_ahah
```

```python
above_avg = ahah[ahah['ahah'] > avg_ahah]
below_avg = ahah[ahah['ahah'] <= avg_ahah]

len(above_avg), len(below_avg)
```

```python
ahah.loc[ahah['ahah'] > avg_ahah, 'average'] = 'above'
ahah.loc[ahah['ahah'] <= avg_ahah, 'average'] = 'below'
```

```python
# Create a unique value map for the labels you have just created

ahah.plot(
    column='average', 
    categorical=True, 
    legend=True
)
```
